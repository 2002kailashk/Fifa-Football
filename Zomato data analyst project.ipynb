```python
import numpy as np
```

```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sn
```

```python
data=pd.read_csv('Zomato-data-.csv')
data
```

```python
data.head()
```

```python
def handleRate(value):
    value=str(value).split('/')
    value=value[0];
    return float(value)

data['rate']=data['rate'].apply(handleRate)
print(data.head())
```

```python
data.info()
```

```python
sn.countplot(x=data['listed_in(type)'])
plt.xlabel("Type of restaurant")
```

```python
grouped_data = data.groupby('listed_in(type)')['votes'].sum()
result = pd.DataFrame({'votes': grouped_data})
plt.plot(result, c='blue', marker='o')
plt.xlabel('Type of restaurant', c='pink', size=20)
plt.ylabel('Votes', c='pink', size=20)
```

```python
sn.countplot(x=data['online_order'])
```

```python
plt.hist(data['rate'],bins=4)
plt.title('Ratings Distribution')
plt.show()
```

```python
couple_data=data['approx_cost(for two people)']
sn.countplot(x=couple_data)
```

```python
plt.figure(figsize = (6,6))
sn.boxplot(x = 'online_order', y = 'rate', data = data)
```

```python
pivot_table = data.pivot_table(index='listed_in(type)', columns='online_order', aggfunc='size', fill_value=0)
sn.heatmap(pivot_table, annot=True, cmap='YlGnBu', fmt='d')
plt.title('Heatmap')
plt.xlabel('Online Order')
plt.ylabel('Listed In (Type)')
plt.show()
```
